Data visualisation assists us with finding patterns and interesting bits of information within datasets that we analyse and to convey these findings with easy to communicate and follow visualisations.

Data scientists perform various functions to analyse data in meaningful ways, however, different people describe these steps in different ways. These steps do not follow a strict order of any sort, and can be describes as such: 
    
    (Data engineering)
        - Data Wrangling
        - Data Clensing
        - Data Preparation

    (Machine Learning)
        - Model Learning
        - Model Validation

    (Operations)
        - Model Deployment
        - Data Visualisation

Data Wrangling can be defined as :  "The process by which you identify, collect, merge, and preprocess one or more datasets in preparation for data clensing" (Jones 2018)

Once we've wrangled that data into a format which we can work with, we clean the data by fixing errors we come across within the dataset. This could potentially involve : 

    - Dealing with missing data
    - correcting formatting issues
    - fixing / removing incorrect data (Depending on severity)

The first job undertaken by a data scientist is to gain a good and clear understanding of the data that they are working with. The data scientist needs to understand what data they have, what format it is in, what format it needs to be converted to, what is wrong with the data and how to fix problems. Various visualisations (such as graphs, tables and charts) are used to gain better insight into the data. 

Therefore, data visualisation is important at both the end of the data science pipeline (presenting the findings of a study) and in te initial phase (data engineering phase) of the data science pipeline to help gain an understanding of data.

Data Visualisation Techniques & Visualising with Python

    Depending on what you would like to find from a dataset, we would need to choose the appropiate visualisations.

    Let us discuss a few basic visualisation techniques : 


```python
'''
Bar Chart - Uses either horizontal or vertical bars to show
    and compare values across categories.

To visualise this with code, we use the Matplotlib Python package
    to allow us to create and display the visualisation.
'''

%pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np

# make dummy data 
x = np.arange(5) # Arange returns evenly spaces values with an interval
y = np.random.uniform(4, 9, len(x)) # Uniform draws samples from a uniform distribution

# Create plot
fig, ax = plt.subplots()

ax.bar(x,y, width=0.7, edgecolor="blue", color="magenta") 
# You can also add extra parameters to customise the looks of the graphs

plt.show()
```

The bar chart is best used for data that is discrete and categorical
    
In data science with often work with different types of statistical data. This data can be numerical, categorical or ordinal

    - Numerical : data can be measured using numbers
    - Categorical : data represents characteristics, such as a persons age
    - Ordinal : both numerical and categorical mixed

Note that numerical can be either : 
    
    - Discrete : data can be counted, such as the number of students in a class
    - Continuous : data cannot be counted, however it can be measured, such as weight, height etc.

```python
'''
Line Graph - shows values over a time period. Multiple line graphs can show and compare many different
    categories over a time period.
        Typically, one axis shows data values, and the other axis depicts time.
'''

import matplotlib.pyplot as plt
import numpy as np

# make dummy data
x = [2009, 2010, 2011, 2012, 2013, 2014]
y = [12.4, 13.9, 11.6, 17.7, 17.1, 18.0]
z = [13.5, 13.7, 12.6, 15.9, 15.7, 17.5]

# create plots
fig, ax = plt.subplots()

# one can create multiple plots on one graph
ax.plot(x, y, color="r")
ax.plot(x,z)

# one can even name the axis with matplotlib
plt.title("INSERT TITLE HERE")
plt.xlabel("Year")
plt.ylabel("Profits")

plt.show()
```

The line graph is best used for data that is linear(continuous) or a time series.
    Keep in mind that time is usually on the x-axis

```python
'''
Pie Chart - great for displaying numerical proportion
'''

import matplotlib.pyplot as plt
import numpy as np

label = ['A', 'B', 'C', 'D']
values = np.linspace(.2,.7, len(label)) # Linspace returns evenly spaced numbers over a set interval

plt.pie(values, labels=label, colors='rgbm')
plt.show()
```

Working with Seaborn

        Seaborn is a data visualisation library that was built on top of matplotlib. While matplotlib is capable of providing basic graphs, such as line and scatter plots, seaborn can provide a bit more in terms of graphing. Additionally, it also integrates nicely with pandas.

For example : let's take data from a dataset for an individuals age and salary, we would like to understand the relationship between these two columns of data. 

Now, with matplotlib, the scatter() method would probably be best for taking a look at the data. 

```python
import matplotlib.pyplot as plt
import numpy as np

# dummy data
age = np.arange(50)
salary = np.random.uniform(20000, 78000, len(age))

plt.scatter(age, salary)
plt.show()
```

We could get a general sense of the data, however it may be difficult to understand the trends immediately. With seaborn, we can use a lineplot() method that automatically plots the averages and standard deviations for ease of reading.

```python
import seaborn as sns

sns.lineplot(x=age, y=salary)
plt.show()
```

This makes it a bit easier to see any trends or pattern within the data.

Some commonly use Seaborn plots would be :

        - histplot()
        - barplot()
        - boxplot()

Let us go over a few more visualisations : 

```python
'''
Stacked area chart - these types of graph are use to compare multiple variables that changes over time.
'''
import matplotlib.pyplot as plt
import numpy as np

# dummy data
groupA = np.random.randint(1,10,10)
groupB = np.random.randint(1,10,10)
groupC = np.random.randint(1,10,10)

# make plot
x = np.arange(10)
y = np.row_stack((groupA,groupB,groupC))

fig, ax = plt.subplots()
ax.stackplot(x,y)

plt.show()
```

In the code above we have created three arrays (groupOne, groupTwo and groupThree) that contain the data that we want to represent on our stacked area
graph. The instruction, y = np.row_stack((groupOne,groupTwo,groupThree)), takes
a sequence of arrays and stacks them vertically to make a single array. The code, x
= np.arange(10), returns an array with evenly spaced elements as per the interval.

With a stacked area graph we want to be able to view different data together in the
same graph. To do this we use subplots. A subplot is defined as, “groups of smaller
axes that can exist together within a single figure. These subplots might be insets,
grids of plots, or other more complicated layouts (VanderPlas, 2016).” Notice that
we use the code, plt.subplots(). This function returns a figure, which is the top
level container for all the plot elements, and an array of Axes objects. The code,
ax.stackplot(x,y), creates a stackplot. Stackplots are generated by plotting
different datasets vertically on top of one another rather than overlapping with one
another.

```python
'''
3D scatterplot - useful for when we would like to compare 3 characteristics of a dataset instead of two. Matplotlib
    is what we will be using to create this scatter plot.
'''

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# create dummy data
n = 10
rng = np.random.rand(n)

xx = (0.1+0.2 * rng, rng, 0.4+0.1 * rng)

yy = (0.7+0.9 * rng, 0.2 * rng, 0.51 * rng)

zz = (0.6 * rng, 0.7 * rng, 0.4+0.1 * rng)

# Create plot
data = (xx, yy, zz)
colours = ('red', 'green', 'blue')
groups = ('A', 'B', 'C')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


for data, colour, group in zip(data, colours, groups):

    x, y, z = data

    ax.scatter(x,y,z , c=colour , alpha=1 , s=35 )

plt.show()

```

From the code above, notice the following:

        - To create 3D visualisations with Matplotlib add the following import
            statement: from mpl_toolkits.mplot3d import Axes3D

        - Once you have added this import statement, the option ‘projection’
            becomes available. By default, Matplotlib produces 2D visualisations. Once
                you have imported mplot3d, you can specify that projection="3d".

        - The code, ax = fig.gca(projection="3d") gets the current 3D axes on the
            figure or creates and returns the appropriate axes if they don’t already exist.

        - ax.scatter() creates a scatterplot.
